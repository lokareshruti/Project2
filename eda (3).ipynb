```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
data=pd.read_csv("LoanApprovalPrediction_cleaned.csv")
data
```

```python
#Remove unnamed column


if 'Unnamed: 0' in data.columns:
    data = data.drop(columns=['Unnamed: 0'])
```

```python
#Target Variables Distribution

plt.figure(figsize=(6,4))
sns.countplot(x='Married',data=data)
plt.title('Loan(Target)')
plt.xlabel('Target (0: No loan,1:loan)')
plt.ylabel('Count')
plt.show()
print(data.columns.tolist())
```

```python
# Education Distribution

plt.figure(figsize=(5, 3))
sns.histplot(data['Education'], kde=True, bins=10)
plt.title('Edicution Distribution')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Distribution

plt.figure(figsize=(5, 3))
sns.histplot(data['LoanAmount'], kde=True, bins=10)
plt.title('LoanAmount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Resting Blood Pressure (trestbps)

plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Property_Area', data=data)
plt.title('LoanAmount vs Property_Area')
plt.show()
```

```python
# Count Plot for Sex

plt.figure(figsize=(6, 4))
sns.countplot(x='Education', data=data, hue='Gender')
plt.title('ApplicantIncome by CoApplicantIncome(0: Female, 1: Male)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
